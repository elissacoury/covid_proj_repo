## Background and Prior Work

During the early stages of the COVID-19 pandemic, Anti-Asian American sentiments surged in the United States, leading to a sense of fear within Asian American communities. ### COVID-19 weekly cases and deaths by US county - CDC

The Weekly United States COVID-19 Cases and Deaths by County dataset from the Centers for Disease Control and Prevention (CDC), is a comprehensive record of reported COVID-19 cases and deaths across the United States. The dataset is organized by week and U.S. jurisdiction. The data were compiled by the CDC and U.S. jurisdictions in a collaborative effort to ensure data accuracy. Throughout the data collection process, the CDC retrospectively updated totals to reflect correct data and address quality issues. 
Important metrics used in this dataset are “cumulative_cases” which provides the total number of confirmed COVID-19 cases reported in a county up to a specific date. “Cumulative_deaths” provides the total number of confirmed COVID-19 reported deaths in a county up to a specific date. “New_cases” refers to the amount of new COVID-19 cases reported in a county during that week. And “new_deaths” is the number of new COVID-19 related deaths that have been reported in a county during that week. Both case and death counts are represented as integer values. Cumulative cases and deaths reflect amounts that have been reported since the start of the pandemic, and new cases and deaths reflect weekly increments of the overall total. For example, within the dataset, Amador County in California was reported to have 1,679 cumulative cases, and 628 new cases on December 16th, 2020. This means that during that week, the total number of cases became 1,679, as 628 new reports were added to that total. These two variables showcase how “new” columns represent weekly change, and “cumulative” columns represent added totals. 
Within the dataset, other metrics include “fips_code” which represents the Federal Information Processing Standard (FIPS) code, providing a unique identifier for each jurisdiction. The “county” and “state” metrics are used to represent a county’s name along with a two letter abbreviation of the state where the county is located within. And “state_fips” refers to the FIPS code that is used to identify and differentiate states. County and state metrics are string/ text based, and use letters in order to be represented.  FIPS related metrics use integers of different lengths in order to differentiate counties and states. 
Some concerns and limitations of this dataset can be apparent through the different ways that counties and states may organize and report their data. As the data relies on indirect county-level counts through either an automated overnight web collection, or a direct submission, inaccuracies may be seen through reporting delays and inconsistency in how cases and deaths are classified across each state and county. Additionally, the CDC also claims that not all U.S. jurisdictions reported probable cases and deaths to the CDC, meaning that discrepancies may exist in the data due to differing COVID-19 surveillance and reconciliation efforts across the country.

###  Twitter Dataset of frequency terms related to COVID-19

This dataset contains 40+ million tweets that relate to COVID-19 and was taken from the platform near the earlier stages of the pandemic. The data was taken from the date January 1, 2020 through March 11, 2020. Phrases and terms relating to COVID-19 are said to be taken from all languages but the most prevalent languages that appeared were English, Spanish, and French.
Some important metrics in this dataset are “counts”, “gram”, and “term”. “Counts” refers to how many times the word(s) appeared in the data, “gram” is just a short hand for either bigram or trigram and it refers to the occurrence of two/three consecutive words. The units these metrics are measured in are in occurrences. For example, gram could refer to bigram count. In the dataset it could show up as a pair of consecutive words like “China virus” and is counted once in the dataset every time it appears. This term could have 2000+ counts over the time period recorded in the dataset which means that there was widespread online discourse regarding racial stigmas and the disease. 
The main concern/limitation of this dataset is the presence of bots or troll accounts on Twitter. Bots may have been used to spam certain terms or phrases, making the figures for a certain term/phrase skew dramatically. Troll accounts also often amplify certain messages or even repost identical content, distorting the actual frequency a term/phrase appears giving the impression that certain sentiments were more prevalent than in reality. Another concern is the contextual situations the terms/phrases were used in. For example, some tweets may have been tweeted in efforts to reprimand using such terms rather than in a harmful manner. Similarly, some people also purposely evade moderation by censoring the term/phrase by using other characters to replace the letters (ex: “ch*nese”/ch1nese”). Additionally, they could have used different keywords instead of censoring the words or outright typing them out. These issues could possibly underestimate the amount of times these terms/phrases appeared. Lastly, since the dataset only includes data from Twitter, it only reflects one social media platform. Different platforms serve different demographics and all use different means of communication. By using only this dataset to represent online hate, it could generalize all online spaces.
