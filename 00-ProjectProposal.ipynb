## Research Question

How did COVID-19 impact the rate of Asian-American hate crimes/sentiment from 2020 to present day? Within this, how did COVID-19 differently impact Asian-American hate online vs. documented incident reports in major U.S. cities? Did they move at the same rate?


## Background and Prior Work

During the early stages of the COVID-19 pandemic, Anti-Asian American sentiments surged in the United States, leading to a sense of fear within Asian American communities. New, derogatory terminology also arose during this time as some referred to the virus as ‘Kung flu” or the “Chinese Virus” <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). Specifically, a study from the University of Utah <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) shows that the distribution of hateful tweets against Asian Americans surged between January and March of 2020 where many began to use harmful language such as calling the virus the “Chinese Virus.”

For instance, survey data from the Pew Research Center illustrated the exact increase of anti-Asian discrimination during the COVID-19 pandemic. Through this, they came to the conclusion that four out of 10 US-born Asian Americans personally know an Asian person who has been attacked or threatened ever since COVID-19 pandemic began <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). Additionally, although xenophobic sentiments have long existed through American history, the emergence of COVID-19 enabled the spread of racism and xenophobia, which may be associated with the increase in anti-Asian hate crime during the pandemic. During the pandemic, many tweets appeared where people stood against the hate crimes against Asian Americans that were recorded and posted. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). Another study conducted looked at hate crime information from police departments from New York, San Francisco, Seattle, and Washington D.C. where they found that New York showed the greatest increase in hate crimes against Asian Americans with an increase of 3200% in 2020 compared to 2019. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). While hate crimes against Asian Americans increased, so did the use of racist terminology.

Although the general impact the pandemic had on anti-Asian sentiments is widely recognized, the direct comparison and degree to which in person and online hate towards Asian Americans is not known precisely. We knew that we wanted to work with Social issues and possible something relating to COVID-19. In the end, we decided upon COVID and Anti-Asian American but were still unclear on how to make our research question more specific. Since our question was definitely not a novel idea, we decided in the end that we wanted to focus on the correlation between COVID-19 and any possible increase in Anti-Asian American by using data from in person hate crime statistics such as targeted robbers, assault, etc. as hateful speech and terminology made online through platforms such as Twitter (X), Instagram, and even news outlets well. Originally, we were considering searching for information in general, urban areas in the United States but ended up deciding to work with cases from major cities in the United States such as Los Angeles, New York, Chicago, etc.
